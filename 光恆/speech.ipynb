```python
import os
import re
import json
data_dir='./data/speech'
md_list=os.listdir(data_dir)
for file in md_list:
    print(file)
```

```python
for file in md_list:
    out=[]
    with open(data_dir+'/'+file) as f:
        title=file[5:-3].strip()
        lines=f.readlines()
        topic=''
        s1=''
        s2=''
        topic=''   
        kp_tag=False
        topic_tag=False 
        for line in lines:
            check_kp=(line.find("#### 柯文哲") != -1)
            check_topic=(line.find("#### ") != -1) and (not line.find("#### tags") != -1) and (not check_kp)
            check_end=(line.find('---')!=-1)
            if(check_topic):
                    topic_tag=True
                    topic=topic+line[5:-1]+':'
                    continue
            
            if(check_kp):
                    kp_tag=True                
                    continue

            if(check_end):
                if(topic_tag):
                    topic_tag=False
                    continue
                else:

                    kp_tag=False
                    topic=''
                    s1=''
                    s2=''
                    continue
                    
            if(topic_tag):
                if(line!='\n'):
                    topic=topic+line

            if(kp_tag):
                if(line!='\n'):
                    tmp=line
                    tmp_list = re.split(r"([。?？])", tmp)
                    #tmp.split('。?')
                    tmp_list.append("")
                    tmp_list = ["".join(i) for i in zip(tmp_list[0::2],tmp_list[1::2])]
                    
                    for sentence in tmp_list:
                        sentence=sentence.strip()
                        if(s1==''):
                            s1=sentence
                        else:
                            s2=sentence
                        if(s2!=''):
                            if(topic==''):
                                topic=title
                            out.append({'topic':topic,'sentence 1':s1,'sentence 2':s2})
                            s1=s2
                    
    file_name=file[:-3].replace('.',' ')
    output_file='./Preprocessed data/speech/'+file_name
    print(output_file)
    with open(output_file, "w") as f:        
        json.dump(out,f,indent=2,ensure_ascii=False)
        
```
